```python
%%sql 

# Table 6

# Assuming user_guid is the primary key for this table, we check if there are any duplicate records for each user_guid
SELECT user_guid, COUNT(*)
FROM ( SELECT DISTINCT * FROM users ) AS distinct_users
GROUP BY user_guid
HAVING COUNT(*) > 1;
```

```python
%%sql

# On further observation, we find that each of these 13 user_guids have 2 corresponding records with only one different value for the field 'utc_correction'
# We consider the informative value for the field 'utc_correction' to be the one which is not '#N/A'
# Based on this, we write the following query : 
# The CTE gives us a list of user_guids that have duplicate records
# We then filter the table with two WHERE clauses - one clause includes all ids not in the invalid list
# the other clause includes ids that are in the invalid list but have the ut

WITH invalid_user_guids AS
( 
SELECT user_guid
FROM ( SELECT DISTINCT * FROM users ) AS distinct_users
GROUP BY user_guid
HAVING COUNT(*) > 1 
) 
SELECT COUNT(*)
FROM ( SELECT DISTINCT * FROM users ) AS distinct_users
WHERE distinct_users.user_guid NOT IN ( SELECT user_guid FROM invalid_user_guids ) 
OR  ( distinct_users.user_guid IN ( SELECT user_guid FROM invalid_user_guids ) AND utc_correction != '#N/A' );
```

```python
%%sql 

# Task 3(a)

WITH invalid_user_guids AS
( 
    SELECT user_guid
    FROM (SELECT DISTINCT * FROM users) AS distinct_users
    GROUP BY user_guid
    HAVING COUNT(*) > 1 
), filtered_users AS
(
    SELECT *
    FROM (SELECT DISTINCT * FROM users) AS distinct_users
    WHERE distinct_users.user_guid NOT IN (SELECT user_guid FROM invalid_user_guids) 
    OR  (distinct_users.user_guid IN (SELECT user_guid FROM invalid_user_guids) AND utc_correction != '#N/A')
)
SELECT 
    YEAR(created_at) AS Year,
    MONTH(created_at) AS Month,
    COUNT(user_guid) AS `signups #`,
    ROUND(100 * SUM(CASE WHEN membership_type = 1 THEN 1 ELSE 0 END) / COUNT(user_guid), 2) AS `membership_type_1 %`,
    ROUND(100 * SUM(CASE WHEN membership_type = 2 THEN 1 ELSE 0 END) / COUNT(user_guid), 2) AS `membership_type_2 %`,
    ROUND(100 * SUM(CASE WHEN membership_type = 3 THEN 1 ELSE 0 END) / COUNT(user_guid), 2) AS `membership_type_3 %`,
    ROUND(100 * SUM(CASE WHEN membership_type = 4 THEN 1 ELSE 0 END) / COUNT(user_guid), 2) AS `membership_type_4 %`,
    ROUND(100 * SUM(CASE WHEN membership_type = 5 THEN 1 ELSE 0 END) / COUNT(user_guid), 2) AS `membership_type_5 %`
FROM 
    filtered_users
GROUP BY 
    YEAR(created_at),
    MONTH(created_at)
ORDER BY 
    Year, 
    Month;
```

```python
%%sql 

# Task 3(b)
WITH invalid_user_guids AS
( 
    SELECT user_guid
    FROM (SELECT DISTINCT * FROM users) AS distinct_users
    GROUP BY user_guid
    HAVING COUNT(*) > 1 
), filtered_users AS
(
    SELECT *
    FROM (SELECT DISTINCT * FROM users) AS distinct_users
    WHERE distinct_users.user_guid NOT IN (SELECT user_guid FROM invalid_user_guids) 
    OR  (distinct_users.user_guid IN (SELECT user_guid FROM invalid_user_guids) AND utc_correction != '#N/A')
)
SELECT     
	YEAR(filtered_users.created_at) AS Year,
    COUNT(filtered_users.user_guid) AS `signups #`,
    SUM(total_tests_completed) AS sum_totaL,
    ROUND(100 * SUM(CASE WHEN membership_type = 1 THEN 1 ELSE 0 END) / COUNT(filtered_users.user_guid), 2) AS `membership_type_1 %`,
	SUM(CASE WHEN membership_type = 1 THEN total_tests_completed ELSE 0 END) AS `tests_completed_1 %`,
    ROUND(100 * SUM(CASE WHEN membership_type = 2 THEN 1 ELSE 0 END) / COUNT(filtered_users.user_guid), 2) AS `membership_type_2 %`,
	SUM(CASE WHEN membership_type = 2 THEN total_tests_completed ELSE 0 END) AS `tests_completed_2 %`,
    ROUND(100 * SUM(CASE WHEN membership_type = 3 THEN 1 ELSE 0 END) / COUNT(filtered_users.user_guid), 2) AS `membership_type_3 %`,
	SUM(CASE WHEN membership_type = 3 THEN total_tests_completed ELSE 0 END) AS `tests_completed_3 %`,
    ROUND(100 * SUM(CASE WHEN membership_type = 4 THEN 1 ELSE 0 END) / COUNT(filtered_users.user_guid), 2) AS `membership_type_4 %`,
	SUM(CASE WHEN membership_type = 4 THEN total_tests_completed ELSE 0 END) AS `tests_completed_4 %`,
    ROUND(100 * SUM(CASE WHEN membership_type = 5 THEN 1 ELSE 0 END) / COUNT(filtered_users.user_guid), 2) AS `membership_type_5 %`,
	SUM(CASE WHEN membership_type = 5 THEN total_tests_completed ELSE 0 END) AS `tests_completed_5 %`
FROM filtered_users
JOIN ( SELECT DISTINCT * FROM dogs ) AS distinct_dogs
ON filtered_users.user_guid = distinct_dogs.user_guid
WHERE filtered_users.user_guid IS NOT NULL AND distinct_dogs.user_guid IS NOT NULL AND distinct_dogs.dog_guid IS NOT NULL
GROUP BY 
    YEAR(filtered_users.created_at);
```
